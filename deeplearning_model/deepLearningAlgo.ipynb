```python
period = '10y'
interval = '1d'

userInputGaran = 'GARAN.IS'
userInputYapiKredi = 'YKBNK.IS'
userInputBIST = 'XU030.IS'
userInputUSD_TRY = 'TRY=X'
userInputJapanese30 = '^N225'

stockHelperData_0 = yfHelper.yFinance(userInputGaran.upper(), period, interval).getData()
stockHelperData_1 = yfHelper.yFinance(userInputYapiKredi.upper(), period, interval).getData()
stockHelperData_2 = yfHelper.yFinance(userInputUSD_TRY.upper(), period, interval).getData()
stockHelperData_3 = yfHelper.yFinance(userInputBIST.upper(), period, interval).getData()
stockHelperData_4 = yfHelper.yFinance(userInputJapanese30.upper(), period, interval).getData()
```

```python
stockHelperData_0 = dpl.obtainLabels(stockHelperData_0)
```

```python
dataList = [stockHelperData_0, stockHelperData_1, stockHelperData_2, stockHelperData_3, stockHelperData_4]
```

```python
for i, df in enumerate(dataList, 0):
    
    for col_name in df.columns:
        
        
        if col_name == "Date":
            pass
        else:
            df.rename(columns = {f"{col_name}" : f"{col_name}" + f"_{i}"}, inplace = True)
            

            
        #df.columns = [col_name+'_{}'.format(i) for col_name in df.columns]
```

```python
merged_df = dataList[0].copy()
for df in dataList[1:]:
    merged_df = pd.merge(merged_df, df, on = 'Date')
```

```python
merged_df
```

```python
import anotherDataProcessLib as adpl
```

```python
scaled_df = adpl.minmaxScaler(merged_df)
```

```python
scaled_df
```

```python
train, test = adpl.train_test_split(scaled_df)
```

```python
n_past = 1
n_future = 1

X_train, y_train = adpl.futureSplitData(train, n_past, n_future, 0)
X_test, y_test = adpl.futureSplitData(test, n_past, n_future, 0)
```

```python
X_train, X_test = adpl.reshapeData_X(X_train, X_test)
y_train, y_test = adpl.reshapeData_y(y_train, y_test)
```

```python
import os
os.environ['KMP_DUPLICATE_LIB_OK']='True'
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
from keras.layers import Masking

model = Sequential()
#model.add(Masking(mask_value=0., input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(LSTM(units = 32, activation = 'relu'))
#model.add(Dropout(.25))
model.add(Dense(1, activation = 'tanh'))
    
model.compile(optimizer='adam', loss = 'mean_squared_error')
model.fit(X_train[:,:,:10], y_train, epochs = 10, batch_size=32)
model.summary()
y_predictions = model.predict(X_test[:,:,:10])
```

```python
import metrics_class as msl
msl.mseCalculator(y_predictions, y_test)
```

```python
from sklearn.metrics import r2_score
print(' R2 degeri:',r2_score(y_test,y_predictions))
```
